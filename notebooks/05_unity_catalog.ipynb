```python
%sql
CREATE CATALOG IF NOT EXISTS cityflow;
CREATE SCHEMA IF NOT EXISTS cityflow.bronze;
CREATE SCHEMA IF NOT EXISTS cityflow.silver;
CREATE SCHEMA IF NOT EXISTS cityflow.gold;
SHOW SCHEMAS IN cityflow;
```

```python
%sql
-- Managed tables  (credentials not required)
CREATE OR REPLACE TABLE cityflow.gold.daily_summary AS
SELECT * FROM cityflow.gold.daily_summary;

CREATE OR REPLACE TABLE cityflow.gold.hourly_demand AS
SELECT * FROM cityflow.gold.hourly_demand;

CREATE OR REPLACE TABLE cityflow.bronze.airquality AS
SELECT * FROM cityflow.bronze.airquality;

SHOW TABLES IN cityflow.gold;
```

```python
%sql
COMMENT ON TABLE cityflow.gold.daily_summary IS 
'NYC Taxi daily trip summary — aggregated from 10.8M cleaned records';

COMMENT ON TABLE cityflow.gold.hourly_demand IS 
'NYC Taxi hourly demand patterns — 168 rows covering 24 hours x 7 days';

COMMENT ON TABLE cityflow.bronze.airquality IS 
'NYC Air Quality hourly data — PM2.5, PM10, Ozone, NO2 from Open-Meteo API';

SELECT COUNT(*) as total_rows, 
       MIN(trip_date) as earliest_date,
       MAX(trip_date) as latest_date,
       ROUND(SUM(total_revenue), 2) as total_revenue
FROM cityflow.gold.daily_summary;
```

```python
%sql
ALTER TABLE cityflow.gold.daily_summary 
ALTER COLUMN total_trips COMMENT 'Total taxi trips for that day';

ALTER TABLE cityflow.gold.daily_summary 
ALTER COLUMN avg_fare COMMENT 'Average fare amount in USD';

ALTER TABLE cityflow.gold.daily_summary 
ALTER COLUMN total_revenue COMMENT 'Total revenue including tips and surcharges';

SHOW TABLES IN cityflow.bronze;
SHOW TABLES IN cityflow.silver;
SHOW TABLES IN cityflow.gold;
```
